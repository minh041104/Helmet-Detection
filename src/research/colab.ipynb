```python
import gradio as gr
import cv2
import tempfile
from roboflow import Roboflow
from PIL import Image, ImageDraw
import numpy as np
import os

# Khởi tạo Roboflow
rf = Roboflow(api_key="TYilqqd4ZPUQ4OVj1Vpt")
model = rf.workspace("minh-rorwz").project("helmet-detection-20241116-fkcqz").version(1).model

# Hàm xử lý ảnh
def process_image(image):
    pil_img = Image.fromarray(image)
    with tempfile.NamedTemporaryFile(suffix=".png", delete=False) as temp_img_file:
        pil_img.save(temp_img_file.name)
        preds = model.predict(temp_img_file.name, confidence=0.2, overlap=30).json()

    draw = ImageDraw.Draw(pil_img)
    for pred in preds["predictions"]:
        x, y = pred["x"], pred["y"]
        w, h = pred["width"], pred["height"]
        label = pred["class"]
        conf = pred["confidence"]

        x1, y1 = x - w / 2, y - h / 2
        x2, y2 = x + w / 2, y + h / 2

        color = "green" if label == "helmet" else "red"
        draw.rectangle([x1, y1, x2, y2], outline=color, width=3)
        draw.text((x1, y1 - 10), f"{label} {conf:.2f}", fill=color)

    return pil_img

# Hàm xử lý video
def process_video(video_path):
    cap = cv2.VideoCapture(video_path)

    fps = cap.get(cv2.CAP_PROP_FPS)
    width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

    temp_output = tempfile.NamedTemporaryFile(suffix=".mp4", delete=False)
    out = cv2.VideoWriter(temp_output.name, cv2.VideoWriter_fourcc(*'mp4v'), fps, (width, height))

    while True:
        ret, frame = cap.read()
        if not ret:
            break

        pil_img = Image.fromarray(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
        with tempfile.NamedTemporaryFile(suffix=".png", delete=False) as temp_img_file:
            pil_img.save(temp_img_file.name)
            preds = model.predict(temp_img_file.name, confidence=0.2, overlap=30).json()

        draw = ImageDraw.Draw(pil_img)
        for pred in preds["predictions"]:
            x, y = pred["x"], pred["y"]
            w, h = pred["width"], pred["height"]
            label = pred["class"]
            conf = pred["confidence"]

            x1, y1 = x - w / 2, y - h / 2
            x2, y2 = x + w / 2, y + h / 2

            color = "green" if label == "helmet" else "red"
            draw.rectangle([x1, y1, x2, y2], outline=color, width=3)
            draw.text((x1, y1 - 10), f"{label} {conf:.2f}", fill=color)

        processed_frame = cv2.cvtColor(np.array(pil_img), cv2.COLOR_RGB2BGR)
        out.write(processed_frame)

    cap.release()
    out.release()
    return temp_output.name

# Giao diện Gradio với 2 tab riêng biệt
with gr.Blocks() as demo:
    gr.Markdown("## 🚧 Helmet Detection for Image and Video")

    with gr.Tab("🖼️ Image"):
        with gr.Row():
            image_input = gr.Image(label="Upload Image")
            image_output = gr.Image(label="Detected Image")
        image_button = gr.Button("Detect Helmet (Image)")
        image_button.click(fn=process_image, inputs=image_input, outputs=image_output)

    with gr.Tab("🎥 Video"):
        with gr.Row():
            video_input = gr.Video(label="Upload Video")
            video_output = gr.Video(label="Detected Video")
        video_button = gr.Button("Detect Helmet (Video)")
        video_button.click(fn=process_video, inputs=video_input, outputs=video_output)

demo.launch()
```

```python
class HelmetDetectionOnly:
    def __init__(self, video_path, output_path, api_key, workspace, workflow_id):
        self.video_path = video_path
        self.output_path = output_path
        self.client = InferenceHTTPClient(api_url="https://serverless.roboflow.com", api_key=api_key)
        self.workspace = workspace
        self.workflow_id = workflow_id

    def run(self):
        cap = cv2.VideoCapture(self.video_path)
        if not cap.isOpened():
            print("Không thể mở video.")
            return

        frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
        frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
        out = cv2.VideoWriter(self.output_path, cv2.VideoWriter_fourcc(*'XVID'), 30, (frame_width, frame_height))

        frame_count = 0
        while True:
            ret, frame = cap.read()
            if not ret:
                break

            temp_frame_path = f"/tmp/frame_{frame_count}.jpg"
            cv2.imwrite(temp_frame_path, frame)

            result = self.client.run_workflow(
                workspace_name=self.workspace,
                workflow_id=self.workflow_id,
                images={"image": temp_frame_path},
                use_cache=True
            )

            model_1_detections = result[0].get('model_1_predictions', [])
            dynamic_crop_data = result[0].get('dynamic_crop', [])

            for det_idx, det in enumerate(model_1_detections):
                # Bounding box gốc (driver) trong ảnh gốc
                parent_prediction = result[0]["predictions"]["predictions"][det_idx]
                parent_x = parent_prediction["x"]
                parent_y = parent_prediction["y"]
                parent_w = parent_prediction["width"]
                parent_h = parent_prediction["height"]

                crop_x1 = int(parent_x - parent_w / 2)
                crop_y1 = int(parent_y - parent_h / 2)

                for pred in det.get("predictions", []):
                    # Tọa độ mũ bảo hiểm trong ảnh crop
                    rel_x1 = int(pred["x"] - pred["width"] / 2)
                    rel_y1 = int(pred["y"] - pred["height"] / 2)
                    rel_x2 = int(pred["x"] + pred["width"] / 2)
                    rel_y2 = int(pred["y"] + pred["height"] / 2)

                    # Chuyển về ảnh gốc bằng cách cộng tọa độ crop
                    x1 = crop_x1 + rel_x1
                    y1 = crop_y1 + rel_y1
                    x2 = crop_x1 + rel_x2
                    y2 = crop_y1 + rel_y2

                    label = pred['class']
                    confidence = pred['confidence']
                    if label == 'helmet':
                        color = (0, 255, 0)
                    elif label == 'no-helmet':
                        color = (0, 0, 255)
                    else:
                        color = (128, 128, 128)

                    cv2.rectangle(frame, (x1, y1), (x2, y2), color, 2)
                    cv2.putText(frame, f"{label} ({confidence:.2f})", (x1, max(0, y1 - 10)),
                                cv2.FONT_HERSHEY_SIMPLEX, 0.6, color, 2)

            os.remove(temp_frame_path)
            out.write(frame)
            frame_count += 1

            # Hiển thị khung hình
            plt.figure(figsize=(10, 6))
            plt.imshow(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
            plt.axis('off')
            plt.title(f"Helmet Detection - Frame {frame_count}")
            plt.show()

        cap.release()
        out.release()
        cv2.destroyAllWindows()
        print(f"Video kết quả đã được lưu tại: {self.output_path}")

if __name__ == "__main__":
    pipeline = HelmetDetectionOnly(
        video_path="/content/istockphoto-501664641-640_adpp_is.mp4",
        output_path="/content/result_video.avi",
        api_key="TYilqqd4ZPUQ4OVj1Vpt",
        workspace="minh-rorwz",
        workflow_id="custom-workflow-4"
    )
    pipeline.run()
```

```python
from inference import InferencePipeline
import cv2

def my_sink(result, video_frame):
    if result.get("output_image"): # Display an image from the workflow response
        cv2.imshow("Workflow Image", result["label_visualization"].numpy_image)
        cv2.waitKey(1)
    print(result) # do something with the predictions of each frame


# initialize a pipeline object
pipeline = InferencePipeline.init_with_workflow(
    api_key="TYilqqd4ZPUQ4OVj1Vpt",
    workspace_name="minh-rorwz",
    workflow_id="custom-workflow-4",
    video_reference="/content/istockphoto-501664641-640_adpp_is.mp4", # Path to video, device id (int, usually 0 for built in webcams), or RTSP stream url
    max_fps=30,
    on_prediction=my_sink
)
pipeline.start() #start the pipeline
pipeline.join() #wait for the pipeline thread to finish
```
